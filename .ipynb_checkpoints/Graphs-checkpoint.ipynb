```python
n_regs = [64, 64, 64, 64 ,64, 64, 64, 64, 64, 64, 64, 64, 64, 64, 64, 64, 256, 256, 256, 256, 256, 256, 256, 256, 256, 256, 256, 256, 256, 256, 256, 256, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024]
rob_entries  = [4 , 4, 4, 4, 16, 16, 16, 16, 64, 64, 64, 64, 256, 256, 256, 256, 4 , 4, 4, 4, 16, 16, 16, 16, 64, 64, 64, 64, 256, 256, 256, 256, 4 , 4, 4, 4, 16, 16, 16, 16, 64, 64, 64, 64, 256, 256, 256, 256]
iq_entries = [4, 16, 64, 256, 4, 16, 64, 256, 4, 16, 64, 256, 4, 16, 64 ,256, 4, 16, 64, 256, 4, 16, 64, 256, 4, 16, 64, 256, 4, 16, 64 ,256, 4, 16, 64, 256, 4, 16, 64, 256, 4, 16, 64, 256, 4, 16, 64 ,256]
```

```python
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3,2)
ax1.scatter(log_n_regs, log_rob_entries)
ax2.scatter(log_n_regs, log_iq_entries)
ax3.scatter(log_n_regs, df_1['miss_rate'])
ax4.scatter(log_n_regs, log_n_regs)
ax5.scatter(log_rob_entries, df_1['miss_rate'])
ax6.scatter(log_rob_entries, df_1['miss_rate'])

    
```

```python
plt.rcParams["figure.figsize"] = [10.00, 7.50]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
img = ax.scatter(rob_entries, iq_entries, n_regs, c=df_1['miss_rate'], cmap='PuBuGn', alpha=1)
plt.colorbar(img)#, cax=cax)
ax.set_xlabel('rob_entries')
ax.set_ylabel('iq_entries')
ax.set_zlabel('n_regs')
ax.set_title("Number of Seconds Simulated")#l2 cache overall miss rate variation")#Number of Seconds Simulated")
plt.savefig("miss_rate.jpg", dpi=300)
#plt.close()
```
